```python
cs_jobs = [
    'Software Engineer', 'Software Developer', 'Software Architect', 'Java Developer',
    'Front-End Developer', 'Front-End Engineer', 'Back-End Developer', 'UI Developer',
    'Data Scientist', 'Data Engineer', 'Data Analyst', 'Web Developer', 'Web Designer',
    'Software Tester', 'QA Analyst', 'QA Engineer', 'Quality Assurance Analyst',
    'Systems Administrator', 'Database Administrator', 'Database Developer',
    'Systems Engineer', 'Systems Analyst', 'Network Engineer', 'Network Administrator',
    'Network Security Specialist', 'Network Technician', 'Network Analyst',
    'IT Support Specialist', 'IT Manager', 'IT Administrator',
    'UX/UI Designer', 'UX Researcher', 'Product Designer', 'Graphic Designer',
    'Digital Marketing Specialist', 'SEO Specialist', 'SEO Analyst', 'SEM Specialist'
]

df_cs = df[df['Job_Title'].isin(cs_jobs)]
```

```python
import pandas as pd
import numpy as np
import re
from collections import Counter

def extract_salary_stats(salary_series):
    """Extract salary statistics handling various formats like $56K, 56K-120K, etc."""
    mins = []
    maxs = []
    
    for salary_str in salary_series:
        if pd.isna(salary_str):
            continue
            
        # Clean the salary string
        salary_clean = str(salary_str).replace('$', '').replace(',', '')
        
        # Handle different patterns
        if 'âˆ’' in salary_clean or '-' in salary_clean:
            # Range format: "56Kâˆ’120K" or "56K-120K"
            parts = re.split('[âˆ’-]', salary_clean)
            if len(parts) == 2:
                try:
                    min_val = int(re.findall(r'\d+', parts[0])[0])
                    max_val = int(re.findall(r'\d+', parts[1])[0])
                    mins.append(min_val)
                    maxs.append(max_val)
                except (IndexError, ValueError):
                    continue
        else:
            # Single value format: "56K"
            try:
                val = int(re.findall(r'\d+', salary_clean)[0])
                mins.append(val)
                maxs.append(val)
            except (IndexError, ValueError):
                continue
    
    if mins and maxs:
        return f"{np.median(mins):.0f}K-{np.median(maxs):.0f}K"
    else:
        return "N/A"

def get_top_benefits(benefits_series):
    """Extract top benefits from benefit strings"""
    all_benefits = []
    for benefit_str in benefits_series:
        if pd.notna(benefit_str) and benefit_str != '':
            # Handle both string and dict-like formats
            benefit_str = str(benefit_str)
            if benefit_str.startswith('{') and benefit_str.endswith('}'):
                # Remove curly braces and quotes
                benefit_str = benefit_str.strip('{}').replace("'", "").replace('"', '')
            
            # Split by comma and clean
            benefits = [b.strip() for b in benefit_str.split(',')]
            all_benefits.extend(benefits)
    
    if all_benefits:
        # Count and get top 5
        benefit_counts = Counter(all_benefits)
        top_benefits = [benefit for benefit, count in benefit_counts.most_common(5)]
        return '; '.join(top_benefits)
    else:
        return "N/A"

def get_experience_range(exp_series):
    """Extract experience range from experience strings"""
    years = []
    for exp_str in exp_series:
        if pd.notna(exp_str):
            # Extract numbers from strings like "2 to 12 Years"
            numbers = re.findall(r'\d+', str(exp_str))
            if numbers:
                years.extend([int(num) for num in numbers])
    
    if years:
        return f"{min(years)}-{max(years)} Years"
    else:
        return "N/A"

def create_role_profiles(df):
    """Create comprehensive role profiles from job data"""
    profiles = []
    
    print(f"Processing {len(df)} rows...")
    print(f"Unique Role+Skills combinations: {df.groupby(['Role', 'skills']).ngroups}")
    
    for i, ((role, skills), group) in enumerate(df.groupby(['Role', 'skills'])):
        if i % 50 == 0:
            print(f"Processed {i} role profiles...")
        
        # Base profile
        profile = {
            'role_id': f"{role}_{hash(skills) % 10000}",
            'Role': role,
            'skills': skills,
            'Job_Description': group['Job_Description'].iloc[0],
            'Responsibilities': group['Responsibilities'].iloc[0],
            'total_postings': len(group),
            'unique_companies': group['Company'].nunique(),
            'unique_portals': group['Job_Portal'].nunique()
        }
        
        # Get all work types for this role
        work_types = group['Work_Type'].unique()
        
        # Process each work type
        for wt in work_types:
            wt_data = group[group['Work_Type'] == wt]
            wt_clean = wt.replace(' ', '_').replace('-', '_')
            
            if wt.lower() == 'intern':
                # Handle internships separately
                profile.update({
                    f'{wt_clean}_companies': '; '.join(wt_data['Company'].unique()[:8]),
                    f'{wt_clean}_salary_range': extract_salary_stats(wt_data['Salary_Range']),
                    f'{wt_clean}_qualifications': '; '.join(wt_data['Qualifications'].unique()),
                    f'{wt_clean}_count': len(wt_data),
                    f'{wt_clean}_benefits': get_top_benefits(wt_data['Benefits']),
                    f'{wt_clean}_portals': '; '.join(wt_data['Job_Portal'].unique())
                })
            else:
                # Handle regular jobs with experience
                profile.update({
                    f'{wt_clean}_companies': '; '.join(wt_data['Company'].unique()[:8]),
                    f'{wt_clean}_experience_range': get_experience_range(wt_data['Experience']),
                    f'{wt_clean}_salary_range': extract_salary_stats(wt_data['Salary_Range']),
                    f'{wt_clean}_qualifications': '; '.join(wt_data['Qualifications'].unique()),
                    f'{wt_clean}_count': len(wt_data),
                    f'{wt_clean}_benefits': get_top_benefits(wt_data['Benefits']),
                    f'{wt_clean}_portals': '; '.join(wt_data['Job_Portal'].unique())
                })
        
        profiles.append(profile)
    
    print(f"Created {len(profiles)} role profiles!")
    return pd.DataFrame(profiles)
```

```python
# Then create the role profiles
profiles = create_role_profiles(df_cs)
print(f"Original dataset: {df_cs.shape}")
print(f"Role profiles: {profiles.shape}")
```
